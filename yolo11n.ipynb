```python
from ultralytics import YOLO
```

# YOLO11 Detection 

```python
# Initiate a pretrained YOLO11 model
model = YOLO("yolo11n.pt")

# Train the model using the 'coco8.yaml' dataset for 100 epochs
results = model.train(data="coco8.yaml", epochs=50, imgsz=640)
```

Test inference

```python
results = model("test/jakarta-traffic.jpg")
```

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
annotated_frame = results[0].plot()  # Annotate the first result
cv2.imwrite("annotated_output-3.jpg", annotated_frame)

image = cv2.imread("annotated_output-3.jpg")
```

```python
plt.imshow(image)
```

# YOLO11 Segmentation

```python
# Initiate a pretrained YOLO11 model
model = YOLO("yolo11n-seg.pt")

# Train the model using the 'coco8.yaml' dataset for 100 epochs
results = model.train(data="coco8-seg.yaml", epochs=50, imgsz=640)
```

```python
results = model("test/jakarta-traffic.jpg")
```

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
annotated_frame = results[0].plot()  # Annotate the first result
cv2.imwrite("annotated_output-2.jpg", annotated_frame)
```

```python
image = cv2.imread("annotated_output-2.jpg")
```

```python
plt.imshow(image)
```
